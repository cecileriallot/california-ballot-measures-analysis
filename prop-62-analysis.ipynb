# Proposition 62 (Repeal of the Death Penalty) Analysis

#### Knight Center for Journalism in the Americas's online course __["Python for Data Journalists: Analyzing Money in Politics"](https://journalismcourses.org/PDJ0517.html)__ by Ben Welsh, Los Angeles Times Data Desk Editor.

Proposition 62 (Repeal of the Death Penalty) was among the 17 ballot measures filed with the State of California and was defeated in the November 2016 elections.

This analysis was conducted using data from the __[California Civic Data Coalition](https://www.californiacivicdata.org/)__.


```python
import pandas as pd
```

Read in the list of propositions and committees from CSV.

```python
props = pd.read_csv("http://www.firstpythonnotebook.org/_static/committees.csv")
```

```python
props.head()
```

```python
props.info()
```

Read in the list of contributions to ballot measures from CSV.

```python
contribs = pd.read_csv("http://www.firstpythonnotebook.org/_static/contributions.csv")
```

```python
contribs.head()
```

```python
contribs.info()
```

Read in the names of the 17 propositions on the ballot in November 2016 to analyze which proposition has the most committees.

```python
props.prop_name.value_counts().reset_index()
```

Show the full name of propositions.

```python
props.prop_name.value_counts()
```

Filter the committees supporting and opposing Proposition 62 (Repeal of Death Penalty).

```python
prop62 = props[props.prop_name == "PROPOSITION 062- DEATH PENALTY. INITIATIVE STATUTE."]
```

Print out the five first rows. 

```python
prop62.head()
```

```python
prop62.info()
```

```python
contribs.info()
```

Filter the contributions made to committees supporting and opposing Proposition 62.

```python
merged = pd.merge(prop62, contribs, on="calaccess_committee_id")
```

```python
merged.head()
```

```python
merged.info()
```

Calculate the total amount of money that was raised by large donors ($100 or above in the state of California) both for and against Proposition 62.

```python
merged.amount.sum()
```

Show the number of contributions made to committees in support and in opposition to Proposition 62.

```python
merged.committee_position.value_counts()
```

Filter the contributions made to committees supporting Proposition 62.

```python
support = merged[merged.committee_position == "SUPPORT"]
```

Filter the contributions made to committees opposing Proposition 62.

```python
oppose = merged[merged.committee_position == "OPPOSE"]
```

Calculate the amount of money that was raised by large donors in support of Proposition 62.

```python
support.amount.sum()
```

Calculate the amount of money that was raised by large donors in opposition to Proposition 62.

```python
oppose.amount.sum()
```

Calculate the percentage of money raised by large donors in support of Proposition 62.

```python
support.amount.sum() / merged.amount.sum()
```

Calculate the percentage of money raised by large donors in opposition to Proposition 62.

```python
oppose.amount.sum() / merged.amount.sum()
```

Show the names of committees supporting Proposition 62.

```python
support.committee_name_x.value_counts()
```

Show the names of committees opposing Proposition 62.

```python
oppose.committee_name_x.value_counts()
```

Show the top ten largest individual contributions to both committees supporting and opposing Proposition 62.

```python
merged.sort_values("amount", ascending=False).head(10)
```

Show the top ten largest individual contributions made to committees opposing Proposition 62.

```python
oppose.sort_values("amount", ascending=False).head(10)
```

Show which committees supporting and opposing Proposition 62 raised the most money from large donors.

```python
merged.groupby("committee_name_x").amount.sum().reset_index().sort_values("amount", ascending=False)
```

Show who among the largest donors gave the most money both for and against Proposition 62.

```python
merged.groupby(["contributor_firstname", "contributor_lastname", "committee_position"]).amount.sum().reset_index().sort_values("amount", ascending=False)
```

Show the number of contributions by state made by large donors to committees supporting Proposition 62.

```python
support.contributor_state.value_counts().reset_index()
```

Show the number of contributions by state made by large donors to committees opposing Proposition 62.

```python
oppose.contributor_state.value_counts().reset_index()
```

Show the amount of contributions by state made by large donors to committees supporting Proposition 62.

```python
support.groupby(["contributor_state", "committee_position"]).amount.sum().reset_index().sort_values("amount", ascending=False)
```

Show the amount of contributions by state made by large donors to committees opposing Proposition 62.

```python
oppose.groupby(["contributor_state", "committee_position"]).amount.sum().reset_index().sort_values("amount", ascending=False)
```

Show the top 30 largest individual contributions made by large donors to committees supporting and opposing Proposition 62 from states other than California.

```python
statecontribs = merged.groupby(["contributor_firstname", "contributor_lastname", "contributor_employer", "contributor_occupation", "contributor_state", "committee_position"]).amount.sum().reset_index().sort_values("amount", ascending=False)
```
